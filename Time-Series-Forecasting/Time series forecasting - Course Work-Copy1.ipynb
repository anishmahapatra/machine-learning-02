```python
# Initial setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
## import timeseries data
data = pd.read_csv('Metrail.csv')

# data.columns = ['Month', 'Passengers']
data['Month'] = pd.to_datetime(data['Month'], format = '%b-%y')
data = data.set_index('Month')
```

```python
data.head(12)
```

```python
# Time series analysis
# Plot timeseries data

data.plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Travellers Traffic Data')
plt.show(block  = False)
```

```python
data['Last_Fill_Travellers'] = data.Travellers.interpolate( method = 'linear')
```

```python
data.to_csv('temp.csv')
```

#### Missing Value Treatment

```python
# Mean Imputation
data = data.assign(Passengers_Mean_Imputation = data.Passengers.fillna(data.Passengers.mean()))
```

```python
# Time series analysis
# Plot timeseries data

data.plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic Data')
plt.show(block  = False)
```

## Linear Interpolation

```python
# Mean Imputation
data = data.assign(Passengers_Mean_Interpolation = data.Passengers.interpolate( method = 'linear'))
```

```python
# Time series analysis
# Plot timeseries data

data[['Passengers_Mean_Interpolation']].plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic Data')
plt.show(block  = False)
```

```python
data['Passengers'] = data['Passengers_Mean_Interpolation']
```
